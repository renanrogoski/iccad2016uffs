```python
import pandas as pd
import csv
import matplotlib.pyplot as plt
headers = ['Case','Nr Entradas', 'Accuracy']
df = pd.read_csv('entradas-vs-accuracy.csv', encoding='utf8', names=headers)

x = df['Case']
data1 = df['Nr Entradas']
data2 = df['Accuracy']
#data3 = df['Lines']

my_dpi=250
fig = plt.figure(figsize=(1200/my_dpi, 800/my_dpi), dpi=my_dpi)
fig, ax1 = plt.subplots()

plt.minorticks_on()
plt.grid(which='major', linestyle='--')
plt.grid(which='minor', linestyle=':')

ax1.set_ylabel('Nr Entradas', color='g')#vertical
ax1.tick_params('y', colors='g')

lns1 = plt.plot(x,data1, 'go', label='Nr Entradas') # green bolinha
lns1 = plt.plot(x,data1, 'k:', color='orange')

#plt.gcf().autofmt_xdate()
ax2 = ax1.twinx()

lns2 = ax2.plot(x,data2, 'r^' , label='Accuracy Mdio') # red triangulo 
lns2 = ax2.plot(x,data2, 'k--', color='blue')  # linha tracejada azul 

ax2.set_ylabel('Acc Mdio %', color='r')
ax2.tick_params('y', colors='r')

ax1.legend(loc='upper left', bbox_to_anchor=(0.1, -0.15), shadow=False, ncol=5)
ax2.legend(loc='upper left', bbox_to_anchor=(0.6, -0.15), shadow=False, ncol=5)
fig.savefig('Entradas-VS-Accuracy.png', dpi=100)
```
