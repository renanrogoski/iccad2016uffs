```python
import pandas as pd
import csv
import matplotlib.pyplot as plt
headers = ['Case','Nr Imputs','Acc']
df = pd.read_csv('Resultados-Plotar.csv', encoding='utf8', names=headers)

x = df['Case']
data1 = df['Nr Imputs']
data2 = df['Acc']
#data3 = df['Lines']

plt.plot( x, data1, 'go', label='Número de Entradas') # green bolinha
plt.plot( x, data1, 'k:', color='orange') # linha pontilha orange

plt.plot( x, data2, 'r^' , label='Accuracy Médio') # red triangulo
plt.plot( x, data2, 'k--', color='blue')  # linha tracejada azul

#plt.plot( x, data3, 'b^') # blue triangulo
#plt.plot( x, data3, 'k--', color='blue')  # linha tracejada azul

plt.axis([0, 60, 0, 300])
plt.title("Mais incrementado")
plt.legend(loc='upper left')
plt.grid(True)
plt.xlabel("Caso de teste")
plt.ylabel("entradas x accurracy")
plt.savefig('Entradas-VS-Accuracy.png')
plt.show()
```
